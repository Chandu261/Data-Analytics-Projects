```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sales_Analysis.csv')
```

# Converting the date column to datetime

```python
df['Date'] = pd.to_datetime(df['Date'])
```

# 1. Total Sales Amount

```python
total_sales = df['Sales_Amount'].sum()
print(f"Total Sales Amount: ${total_sales:,.2f}")
```

# 2. Total Quantity Sold

```python
total_quantity = df['Quantity_Sold'].sum()
print(f"Total Quantity Sold: {total_quantity:,}")
```

# 3. Total Profit (assuming 20% profit margin)

```python
df['Profit'] = df['Sales_Amount'] * 0.2
total_profit = df['Profit'].sum()
print(f"Total Profit: ${total_profit:,.2f}")
```

# 4. Average Sales Amount per Day

```python
avg_sales_per_day = df.groupby('Date')['Sales_Amount'].sum().mean()
print(f"Average Sales Amount per Day: ${avg_sales_per_day:,.2f}")
```

# 6. Profit Margin

```python
profit_margin = (total_profit / total_sales) * 100
print(f"Profit Margin: {profit_margin:.2f}%")
```

# **Trend** **Visualizations:**

```python
# Set up the plot style
plt.style.use('seaborn')
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# 1. Sales Trend over Time
df.groupby('Date')['Sales_Amount'].sum().plot(ax=axes[0])
axes[0].set_title('Sales Trend over Time')
axes[0].set_ylabel('Sales Amount')

# 2. Profit Trend over Time
df.groupby('Date')['Profit'].sum().plot(ax=axes[1])
axes[1].set_title('Profit Trend over Time')
axes[1].set_ylabel('Profit')

# 3. Quantity Sold Trend over Time
df.groupby('Date')['Quantity_Sold'].sum().plot(ax=axes[2])
axes[2].set_title('Quantity Sold Trend over Time')
axes[2].set_ylabel('Quantity Sold')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
# Group by month for a monthly trend
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales_Amount'].sum()
monthly_profit = df.groupby('Month')['Profit'].sum()
monthly_quantity = df.groupby('Month')['Quantity_Sold'].sum()

# Plot monthly trends
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

monthly_sales.plot(ax=axes[0])
axes[0].set_title('Monthly Sales Trend')
axes[0].set_ylabel('Sales Amount')

monthly_profit.plot(ax=axes[1])
axes[1].set_title('Monthly Profit Trend')
axes[1].set_ylabel('Profit')

monthly_quantity.plot(ax=axes[2])
axes[2].set_title('Monthly Quantity Sold Trend')
axes[2].set_ylabel('Quantity Sold')

plt.tight_layout()
plt.show()
```
